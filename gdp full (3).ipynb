```python
import pandas as pd
import matplotlib.pyplot as plt

# Load both datasets if not already loaded
gdp_df = pd.read_csv(r"G:\Besent Technology\datasets\New folder\World Bank\New folder\cleaned_gdp_data.csv")
country_df = pd.read_csv(r"G:\Besent Technology\datasets\New folder\World Bank\New folder\country.csv")

# Filter GDP data to years 2000–2023
gdp_years = [str(y) for y in range(2000, 2024)]
gdp_subset = gdp_df[['Country Code', 'Country Name'] + gdp_years].copy()

# Merge with region metadata
merged_df = pd.merge(gdp_subset, country_df[['Country Code', 'Region']], on='Country Code', how='left')

# Melt to long format
long_df = pd.melt(merged_df, id_vars=['Country Code', 'Country Name', 'Region'], 
                  value_vars=gdp_years, var_name='Year', value_name='GDP Per Capita')

# Convert GDP to numeric
long_df['GDP Per Capita'] = pd.to_numeric(long_df['GDP Per Capita'], errors='coerce')

# Drop missing values
long_df.dropna(subset=['GDP Per Capita', 'Region'], inplace=True)

# Calculate volatility: standard deviation per region (across countries)
region_volatility = (
    long_df.groupby(['Region', 'Country Name'])['GDP Per Capita']
    .std()
    .groupby('Region')
    .mean()
    .sort_values(ascending=False)
)

# Plot the volatility by region
plt.figure(figsize=(10,6))
region_volatility.plot(kind='bar', color='steelblue')
plt.ylabel('Average Standard Deviation of GDP Per Capita (2000–2023)')
plt.title('Regional GDP Per Capita Volatility (2000–2023)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Save the plot to your local drive
save_path = r"G:\Besent Technology\datasets\New folder\World Bank\volatility in gdp per capita (2023).png"
plt.savefig(save_path)

plt.show()
print(f"Visualization saved to: {save_path}")
```

```python

# Prepare years and merge with country metadata
gdp_years = [str(y) for y in range(2000, 2024)]
gdp_subset = gdp_df[['Country Code', 'Country Name'] + gdp_years]
merged_df = pd.merge(gdp_subset, country_df[['Country Code', 'Region']], on='Country Code', how='left')

# Melt to long format
long_df = pd.melt(merged_df, id_vars=['Country Code', 'Country Name', 'Region'],
                  value_vars=gdp_years, var_name='Year', value_name='GDP Per Capita')

# Clean
long_df['GDP Per Capita'] = pd.to_numeric(long_df['GDP Per Capita'], errors='coerce')
long_df.dropna(subset=['GDP Per Capita', 'Region'], inplace=True)

# Calculate mean and std per country
stats_df = long_df.groupby(['Region', 'Country Name'])['GDP Per Capita'].agg(['mean', 'std']).reset_index()
stats_df['CV (%)'] = (stats_df['std'] / stats_df['mean']) * 100

# Now get average CV per region
region_cv = stats_df.groupby('Region')['CV (%)'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10,6))
region_cv.plot(kind='bar', color='steelblue')
plt.ylabel('Average Coefficient of Variation (%)')
plt.title('Regional GDP Per Capita Volatility (CV, 2000–2023)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Save the plot to your local drive
save_path = r"G:\Besent Technology\datasets\New folder\World Bank\regional cv.png"
plt.savefig(save_path)

plt.show()
print(f"Visualization saved to: {save_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
gdp_df = pd.read_csv(r"G:\Besent Technology\datasets\New folder\World Bank\New folder\cleaned_gdp_data.csv")
country_df = pd.read_csv(r"G:\Besent Technology\datasets\New folder\World Bank\New folder\country.csv")

# Define GDP years
gdp_years = [str(y) for y in range(2000, 2024)]
gdp_subset = gdp_df[['Country Code', 'Country Name'] + gdp_years]

# Merge GDP and metadata
merged_df = pd.merge(gdp_subset, country_df[['Country Code', 'IncomeGroup']], on='Country Code', how='left')

# Melt to long format
long_df = pd.melt(merged_df, id_vars=['Country Code', 'Country Name', 'IncomeGroup'],
                  value_vars=gdp_years, var_name='Year', value_name='GDP Per Capita')
long_df['GDP Per Capita'] = pd.to_numeric(long_df['GDP Per Capita'], errors='coerce')
long_df.dropna(subset=['GDP Per Capita', 'IncomeGroup'], inplace=True)

# Calculate mean and std per country
stats_df = long_df.groupby(['IncomeGroup', 'Country Name'])['GDP Per Capita'].agg(['mean', 'std']).reset_index()
stats_df['CV (%)'] = (stats_df['std'] / stats_df['mean']) * 100

# Add back IncomeGroup for each country
stats_df['IncomeGroup'] = stats_df['IncomeGroup']

# Plot
plt.figure(figsize=(10,6))
sns.stripplot(
    data=stats_df,
    x='IncomeGroup',
    y='CV (%)',
    hue='IncomeGroup',          # Set hue to match x
    palette='Set2',
    jitter=True,
    alpha=0.7,
    size=7,
    legend=False                # Avoid double legend
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
gdp_df = pd.read_csv(r"G:\Besent Technology\datasets\New folder\World Bank\New folder\cleaned_gdp_data.csv")
country_df = pd.read_csv(r"G:\Besent Technology\datasets\New folder\World Bank\New folder\country.csv")

# Merge with IncomeGroup
merged = pd.merge(
    gdp_df,
    country_df[['Country Code', 'IncomeGroup']],
    on='Country Code', how='left'
)

# Reshape GDP data: from wide to long
long_df = pd.melt(
    merged,
    id_vars=['Country Name', 'Country Code', 'IncomeGroup'],
    value_vars=[str(y) for y in range(2000, 2024)],
    var_name='Year',
    value_name='GDP Per Capita'
)

# Clean and convert
long_df['GDP Per Capita'] = pd.to_numeric(long_df['GDP Per Capita'], errors='coerce')
long_df.dropna(subset=['GDP Per Capita', 'IncomeGroup'], inplace=True)

# Convert year to int
long_df['Year'] = long_df['Year'].astype(int)

# Group by IncomeGroup and Year
income_trend = long_df.groupby(['IncomeGroup', 'Year'])['GDP Per Capita'].mean().reset_index()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=income_trend, x='Year', y='GDP Per Capita', hue='IncomeGroup', marker='o', palette='Set1')

plt.title('GDP Per Capita Trend by Income Group (2000–2023)')
plt.ylabel('Average GDP Per Capita')
plt.xlabel('Year')
plt.grid(True)
plt.legend(title='Income Group')
plt.tight_layout()
plt.show()
save_path = r"G:\Besent Technology\datasets\New folder\World Bank\GDP Per Capita Trend by Income Group.png"
plt.savefig(save_path)

plt.show()
print(f"Visualization saved to: {save_path}")
```
